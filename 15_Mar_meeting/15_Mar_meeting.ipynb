### (2) Simulation population

```python
# Create an empty dictionary to store population
country_population = {}

# Iterate through the data and organize it according to country, subdivision, and commune structure
for index, row in population.iterrows():
    country = row["Country"]
    subdivision = row["Subdivision"]
    commune = row["Communes"]
    population_2022 = row["Population (2022)"]

    # If the country is not in the dictionary yet, create a new dictionary entry
    if country not in country_population:
        country_population[country] = {}

    # If the subdivision is not in the country's dictionary yet, create a new dictionary entry
    if subdivision not in country_population[country]:
        country_population[country][subdivision] = {}

    # Add commune information to the subdivision's dictionary
    country_population[country][subdivision][commune] = population_2022

# Print the organized dictionary
country_population
```

### (4) Simulation Connections depending on airline and marine

```python
# Save routes in a dictionary
connections = {}

airline = air
marine = sea

# Map airport code with locations
name_mapping = {}
for index, row in airline.iterrows():
    name_mapping[row["Depart"]] = row["Airport"]

# åˆ›å»ºä¸€ä¸ªç©ºå­—å…¸æ¥å­˜å‚¨æœºåœºé™„è¿‘çš„ç¤¾åŒºä¿¡æ¯
airportline = {}

# éåŽ†æœºåœºæ•°æ®å¹¶æž„å»ºå­—å…¸
for index, row in airline.iterrows():
    depart = row["Depart"]
    arrival = row["Arrival"].split(",")
    airportline[name_mapping[depart]] = [name_mapping[arr] for arr in arrival if arr in name_mapping]

searoute = {}

for index, row in marine.iterrows():
    depart = row["Depart"]
    arrival = row["Arrival"].split(",")
    searoute[depart] = arrival
```

```python
# Map airport code with locations
name_mapping = {}
for index, row in air.iterrows():
    name_mapping[row["Depart"]] = row["Airport"]

# åˆ›å»ºä¸€ä¸ªç©ºå­—å…¸æ¥å­˜å‚¨æœºåœºé™„è¿‘çš„ç¤¾åŒºä¿¡æ¯
airline_routes = {}

# éåŽ†æœºåœºæ•°æ®å¹¶æž„å»ºå­—å…¸
for index, row in air.iterrows():
    depart = row["Depart"]
    arrival = row["Arrival"].split(",")
    airline_routes[name_mapping[depart]] = [name_mapping[arr] for arr in arrival if arr in name_mapping]

marine_routes = {}

for index, row in sea.iterrows():
    depart = row["Depart"]
    arrival = row["Arrival"].split(",")
    marine_routes[depart] = arrival
```

```python
def create_connectivity_matrix(D, L, location_names):
    W = np.zeros((D, D), dtype=int)
    
    # Populate W with random values and ensure symmetry
    for i in range(1, D):
        for j in range(i+1, D):
            W[i, j] = np.random.randint(0, 100)  # Adjust range as needed
            W[j, i] = W[i, j]

    # Set the connections from the outside world
    for i in range(1, L+1):
        W[0, i] = W[i, 0] = np.random.randint(1, 100)  # Ensure these are non-zero

    # Zero out the diagonal
    np.fill_diagonal(W, 0)
    
    # Convert the matrix to a DataFrame with location names as index and columns
    W_df = pd.DataFrame(W, columns=location_names, index=location_names)
    return W_df

# Number of locations (excluding outside world)
D = len(locations_population)

# Number of connections to the outside world
L = len(airport_communities['Tahiti'])

# Location names
location_names = [location for location, _ in locations_population]

# Create the connectivity matrix DataFrame
W_df = create_connectivity_matrix(D, L, location_names)

# Print the DataFrame
W_df
```

```python
# Calculate the first population quantity for the outside world (arrivals)
arrivals = (236.6 + 77) * 1000  # Sum of 2019 arrivals and 2020 arrivals
outside = ['arrivals',arrivals]
# Convert locations_population to the required format
n = np.array([pop for loc, pop in locations_population])
n = np.insert(n, 0, arrivals)  # Insert arrivals as the first population quantity
n
```

```python
# Initialize state vectors
x = np.zeros((D, 4), dtype=float)  # Use float type for the state vectors

for i in range(D):
    if n[i] == 0:
        x[i] = np.zeros(4)  # Explicitly set to zero for locations with zero population
    else:
        # Directly assign the percentages
        percentages = np.array([0.80, 0.05, 0.05, 0.10])  # Susceptible, Exposed, Infected, Recovered
        x[i] = percentages * n[i]  # Scale by population
        x[i] = np.round(x[i]).astype(int)  # Round to integer

# Convert the state vectors to a DataFrame with location names as index and columns
state_vectors_df = pd.DataFrame(x, columns=['Susceptible', 'Exposed', 'Infected', 'Recovered'], 
                                index=location_names)
state_vectors_df
```
