```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
outside_world = pd.read_excel("Arrivals by region.xlsx")

# Replace 'NaN' and placeholders with 0
outside_world = outside_world.fillna(0)
outside_world = outside_world.replace('..', 0)


outside_world
```

```python
# Ensure the 'Country' column is a string to prevent errors during filtering
outside_world['Country'] = outside_world['Country'].astype(str)

# Filter the DataFrame for 'French Polynesia'
fp_data = outside_world[outside_world['Country'] == 'SAMOA']

# Plot the data with annotations
plt.figure(figsize=(14, 7))

for i in range(len(fp_data)):  # Iterate over the rows for each region within French Polynesia
    year_columns = outside_world.columns[5:]
    arrivals = fp_data.iloc[i, 5:]
    plt.plot(year_columns, arrivals, marker='o', label=fp_data.iloc[i]['Arrivals by region'])
    
    # Add annotations for each point
    for x, y in zip(year_columns, arrivals):
        if y != 0:  # Avoid cluttering the plot with zeros
            plt.text(x, y, f'{y:.0f}', fontsize=8, rotation=45, ha='right', va='bottom')

plt.title('Tourist Arrivals to Samoa by Region Over the Years')
plt.xlabel('Year')
plt.ylabel('Arrivals (Thousands)')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='best')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

# Parameters
Tmax = 60
D = len(nList)   # Number of locations
L = 1   # Number of locations connected to the outside world
n = nList# Population at each location

# Check outside population
outside_population = last / 365 * Tmax
outside_population
```

```python
import numpy as np
import pandas as pd

def create_connectivity_matrix(D, L, outside_population):
    np.random.seed(42)  # Set seed for reproducibility, adjust as needed
    W = np.zeros((D, D), dtype=int)

    for i in range(1, D):
        for j in range(i + 1, D):
            W[i, j] = np.random.randint(0, 100)  # Adjust range as needed
            W[j, i] = W[i, j]

    # Set the connections from the outside world using outside_population array
    for i in range(1, L + 1):
        W[0, i] = W[i, 0] = outside_population

    # Zero out the diagonal
    np.fill_diagonal(W, 0)

    return W

def display_results(W, nList):
    W_df = pd.DataFrame(W, columns=[f"Location {i + 1}" for i in range(D)], index=[f"Location {i + 1}" for i in range(D)])
    print("Connectivity Matrix:")
    print(W_df)
    print("Population Array:")
    print(nList)

# Create connectivity matrix
W_matrix = create_connectivity_matrix(D, L, outside_population)

# Update the population array
nList[0] = outside_population

# Display results
display_results(W_matrix, nList)
```

```python
# Initialize state vectors
x = np.zeros((D, 4), dtype=float)  # Use float type for the state vectors

for i in range(D):
    if n[i] == 0:
        x[i] = np.zeros(4)  # Explicitly set to zero for locations with zero population
    else:
        # Directly assign the percentages
        percentages = np.array([0.00, 0.01, 0.00, 0.00])  # Susceptible, Exposed, Infected, Recovered
        x[i] = percentages * n[i]  # Scale by population
        x[i] = np.round(x[i]).astype(int)  # Round to integer
```

```python
# Convert the state vectors to a DataFrame for better visualization
state_vectors_df = pd.DataFrame(x, columns=['Susceptible', 'Exposed', 'Infected', 'Recovered'], 
                                index=[f"Location {i+1}" for i in range(D)])
state_vectors_df
```
