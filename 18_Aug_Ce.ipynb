# The SIR Model

Predicting the number of Covid-19 infections with SIR Model

Singh, R., Lal, R., & Kotti, R. (2022). Time-discrete SIR model for COVID-19 in Fiji. Epidemiology & Infection, 150, E75. doi:10.1017/S0950268822000590

https://www.cambridge.org/core/journals/epidemiology-and-infection/article/timediscrete-sir-model-for-covid19-in-fiji/3BD9D38A3C0ED24BC409FB7DBF66E82D

S: The number of susceptible people.

α: Transmission rate.

I: The number of infections individuals.

β: Recovering or death rate.

R: The number of removed individuals.

N: The total population

$$
N = S + I + R
$$

Susceptible individuals' rate at time t:
$$
\frac{dS}{dt} = -\alpha \cdot \frac{S \cdot I}{N}
$$



Infectious individuals' rate at time t:
$$
\frac{dI}{dt} = \alpha \cdot \frac{S \cdot I}{N} - \beta \cdot I
$$


Removed or Recovered individuals' rate at time t:
$$
\frac{dR}{dt} = \beta \cdot I
$$


```python
import scipy.integrate
import numpy as np
import matplotlib.pyplot as plt

#SIR Model
def sir_model(y, t, beta, gamma):
    S, I, R = y
    dS_dt = -alpha * S * I / N
    dI_dt = alpha * S * I / N - beta * I
    dR_dt = beta * I
    return ([dS_dt, dI_dt, dR_dt])
```

The basic reproductive number $R_0$(R naught)：

$$R_0= \frac{α}{β}$$

Where,α is the transmission rate and β is the recovery rate.


The basic reproductive number is the number of people who contract the disease from an infected person assuming the whole population is susceptible.

If $R_0 < 1$, Infectious diseases will gradually disappear.

If $R_0 = 1$, Infectious diseases become endemic.

If $R_0 > 1$, Infectious diseases spread exponentially and become epidemics.

So, the larger the number of R0, the more difficult it is to control the epidemic.

```python
# Parameters
alpha = 2.7818 * 0.0403  # Infection rate
beta = 0.0403  # Recovery rate
```

```python
# Initial conditions
N =  924610 # Total population in Fiji（2021）
I0 = 1    # Initial number of infectious individuals
R0 = 0    # Initial number of recovered individuals
S0 = N - I0 - R0  # Initial number of susceptible individuals
```

```python
# Time vector
t = np.linspace(0, 1000, 10000)

# Result
res = scipy.integrate.odeint(sir_model, [S0, I0, R0], t, args=(alpha, beta))
res = np.array(res)

# Plot
plt.figure(figsize=[6, 4])
plt.plot(t, res[:, 0], label='S(t)')
plt.plot(t, res[:, 1], label='I(t)')
plt.plot(t, res[:, 2], label='R(t)')
plt.legend()
plt.grid()
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('SIR Model Simulation in Fiji')
plt.show()
```
