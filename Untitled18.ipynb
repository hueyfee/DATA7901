```python
import numpy as np
import matplotlib.pyplot as plt

class ExtendedSEIRModel:
    def __init__(self, n, initial_population, timesteps, beta, sigma, gamma, transport_rate=0.01,
                 lockdown_start_day=None, lockdown_effectiveness=None):
        self.n = n
        self.initial_population = initial_population
        self.timesteps = timesteps
        self.beta = beta
        self.sigma = sigma
        self.gamma = gamma
        self.transport_rate = transport_rate
        self.lockdown_start_day = lockdown_start_day
        self.lockdown_effectiveness = lockdown_effectiveness

        # Initialize arrays to store SEIR scenarios, population, and transport
        self.seir_matrix = np.zeros((4, n, timesteps))
        self.population_matrix = np.zeros((1, n))
        self.transport_matrix = np.zeros((n, n, timesteps))  # To store the number of people moving between islands

    def initialize(self, initial_infected):
        self.seir_matrix[2, :, 0] = initial_infected
        self.population_matrix[0] = self.initial_population

    def simulate(self):
        for t in range(1, self.timesteps):
            for island in range(self.n):
                S, E, I, R = self.seir_matrix[:, island, t - 1]
                transport_from_other_islands = self.transport_rate * np.sum([
                    self.seir_matrix[0, other_island, t - 1] for other_island in range(self.n) if other_island != island])

                # Calculate transport between islands
                for other_island in range(self.n):
                    if other_island != island:
                        transport_amount = self.transport_rate * self.seir_matrix[0, island, t - 1]
                        self.transport_matrix[island, other_island, t] = transport_amount

                        # Migration step
                        X = np.concatenate([np.repeat("S", int(S)),
                                            np.repeat("E", int(E)),
                                            np.repeat("I", int(I)),
                                            np.repeat("R", int(R))])

                        for compartment in range(4):
                            # Sample without replacement
                            sample_size = int(self.transport_matrix[island, other_island, t])
                            sampled_indices = np.random.choice(len(X), size=sample_size, replace=False)

                            # Update the migration vectors
                            count_S = np.sum(X[sampled_indices] == 'S') if compartment == 0 else 0
                            count_E = np.sum(X[sampled_indices] == 'E') if compartment == 1 else 0
                            count_I = np.sum(X[sampled_indices] == 'I') if compartment == 2 else 0
                            count_R = np.sum(X[sampled_indices] == 'R') if compartment == 3 else 0

                            self.transport_matrix[island, other_island, t] += count_S + count_E + count_I + count_R

                            # Remove the sampled elements
                            X = np.delete(X, sampled_indices)

                   



                if self.lockdown_start_day is not None and t >= self.lockdown_start_day:
                    beta_eff = self.beta * (1 - self.lockdown_effectiveness)
                else:
                    beta_eff = self.beta

                dS = -beta_eff * S * I / self.initial_population[island] + transport_from_other_islands
                dE = beta_eff * S * I / self.initial_population[island] - self.sigma * E
                dI = self.sigma * E - self.gamma * I
                dR = self.gamma * I - transport_from_other_islands

                S += dS
                E += dE
                I += dI
                R += dR

                S = max(S, 0)
                E = max(E, 0)
                I = max(I, 0)
                R = max(R, 0)

                self.seir_matrix[:, island, t] = [S, E, I, R]

    def plot(self):
        plt.figure(figsize=(12, 8))
        days = range(self.timesteps)

        for island in range(self.n):
            plt.subplot(4, self.n, island + 1)
            plt.plot(days, self.seir_matrix[0, island, :], label="S")
            plt.title(f'Island {island + 1} - Susceptible')

            plt.subplot(4, self.n, self.n + island + 1)
            plt.plot(days, self.seir_matrix[1, island, :], label="E")
            plt.title(f'Island {island + 1} - Exposed')

            plt.subplot(4, self.n, 2 * self.n + island + 1)
            plt.plot(days, self.seir_matrix[2, island, :], label="I")
            plt.title(f'Island {island + 1} - Infectious')

            plt.subplot(4, self.n, 3 * self.n + island + 1)
            plt.plot(days, self.seir_matrix[3, island, :], label="R")
            plt.title(f'Island {island + 1} - Recovered')

        for i in range(4 * self.n):
            plt.subplot(4, self.n, i + 1)
            plt.legend(loc="upper right")
            plt.xlabel("Days")
            plt.ylabel("Population")

        plt.tight_layout()
        plt.show()

# Example usage:
n = 2  # Number of islands
initial_population = np.array([1000, 800])
timesteps = 365  # Number of time steps for one year
beta = 0.3  # Transmission rate
sigma = 0.1  # Incubation rate
gamma = 0.05  # Recovery rate

model = ExtendedSEIRModel(n, initial_population, timesteps, beta, sigma, gamma)
model.initialize(initial_infected=np.array([5, 3]))
model.simulate()
model.plot()




# def migration_step(state, W, L):
#     """
#     Perform the migration step.

#     Parameters:
#     - state: Current state matrix for all locations
#     - W: Connectivity matrix
#     - L: Number of connections with the outside world

#     Returns:
#     - new_state: Updated state matrix after migration
#     """
#     D = len(state)
#     new_state = np.copy(state)

#     for i in range(2, L + 2):
#         # Perform migration for connections with the outside world
#         departure_vector = sample_departure_vector(state[i - 1])
#         for j in range(1, D + 1):
#             if j != i:
#                 update_state(new_state, i, j, departure_vector, W[i - 1, j - 1])

#     for i in range(L + 2, D + 2):
#         # Perform migration for domestic locations
#         departure_vector = sample_departure_vector(state[i - 1])
#         for j in range(1, D + 1):
#             if j != i:
#                 update_state(new_state, i, j, departure_vector, W[i - 1, j - 1])

#     return new_state
```
